```python
import numpy as np
import pandas as pd
import os
import re
```

```python
DATA_FOLDER = 'data/'
all_data = pd.read_csv(os.path.join(DATA_FOLDER, 'all_data.csv'))
all_data
```

```python
print('Before removing hashtags - ', all_data[all_data['web-scraper-order'] == '1604198713-1098']['text'])
all_data['text'] = all_data['text'].apply(lambda text: re.sub(r'\#\S+', '', text))
print('After removing hashtags - ', all_data[all_data['web-scraper-order'] == '1604198713-1098']['text'])
```

```python
def get_predictions_vocatives(df):
    return 1 - df['userMentions'].isna().astype(int)
```

```python
def get_predictions_group_inclusive_pronouns(df):
    PRONOUNS = ['everybody', 'everyone', 'our', 'ours', 'ourself', 'ourselves', 'us', 'we']
    return df['text'].apply(lambda text: int(any(word in text.split() for word in PRONOUNS)))
```

```python
# category 4 - Continuing a thread
all_data[all_data['isReply'] == 1][['web-scraper-order', 'text']]
```

```python
# category 9 - Vocatives
all_data[get_predictions_vocatives(all_data) == 1][['web-scraper-order', 'text']]
```

```python
# category 10 - Addresses or refers to the group using inclusive pronouns
all_data[get_predictions_group_inclusive_pronouns(all_data) == 1][['web-scraper-order', 'text']]
```
